```python
from faker import Faker
import pandas as pd
import random
import numpy as np
#import itertools
```

```python
fake = Faker()
def create_rows(num=1):
    output = [{"name":fake.name(),
               #"address":fake.address(),
               "name":fake.name(),
               # "email":fake.email(),
               # "bs":fake.bs(),
               # "address":fake.address(),
               "city":fake.city(),
               "state":fake.state(),
               #"date_time":fake.date_time(),

               # "Conrad":fake.catch_phrase(),
               "account":random.randint(1000,1030),
               #"quantity":random.randint(1,10),
               #"price":random.randint(1,10),
              } for x in range(num)]
    return output
```

```python
#%%time
df = pd.DataFrame(create_rows(500))
catList = ['server', 'client', 'hybrid']
cpuList = ['intel', 'amd', 'arm']
bitList = [8,16,32,64]
memList = [1,2,3,4,6,8,12,16,24,32,48,64,128]
#cpuList = ['_'.join(map(str, list(x))) for x in itertools.product(['intel', 'amd', 'arm'], 
#                                                                  [1000, 2000, 3000], 
#                                                                  [16, 32, 64])]
df['category'] = random.choices(catList, k=len(df))
df['cpu'] = random.choices(cpuList, k=len(df))
df['bit'] = random.choices(bitList, k=len(df))

df['GB'] = random.choices(memList, k=len(df))
df['quantity'] = random.choices([x for x in range(10)], k=len(df))
# [x for x in random.randint(1,10) range(len(df))]
df['price'] = [int(random.expovariate(1/1000))+1 for i in range(len(df))]
```

```python
df.head()
```

## Sum prices by State and CPU

```python
pd.pivot_table(df,index=["state","cpu"],values=["price"],aggfunc=np.sum)
```

## Sum quantity  by Category and CPU

```python
pd.pivot_table(df,index=["category","cpu"],values=["quantity"],
               aggfunc=[np.sum],fill_value=0,margins=True)
```

## Sum and Count of Bit Types by Account and CPU

```python
pd.pivot_table(df,index=["account","cpu"],columns=["bit"],values=["quantity","price"],
               aggfunc={"quantity":len,"price":np.sum},fill_value=0)
```
